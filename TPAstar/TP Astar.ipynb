## Heursitiques
### Parmis ces heuristiques, lesquelles sont admissibles ?
> h0, h1, h2 et h3 sont admissibles car toutes ces distances sont forcément plus petites ou égales à la vraie distance. La distance de Manhattan (h4) en revanche peut être égale à la vraie distance uniquement si le voyage est purement vertical ou horizontal, ce qui n'est qu'un cas particulier improbable.



### Dans un cas réel, quelle heuristique utiliseriez vous ?
> Ici, l'heuristique **h3** (vol d'oiseau) est la meilleure car elle estime le mieux la distance réele, légèrement optimiste ce qui est très bon. Les heuristiques h0, h1 et h2 sont optimiste mais se rapporche moins de la vérité de manière générale, et l'heuristique h4 n'est pas optimiste donc fausse.
